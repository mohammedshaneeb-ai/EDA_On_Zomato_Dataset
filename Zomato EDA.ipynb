## Pie Chart- Top 3 countries that uses zomato


```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
country_val = final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.3f%%')
```

### Observation:
    ####Zomato maximum records or transaction are from India After that USA and then United Kingdoms



```python
final_df.columns
```

```python
final_df[['Aggregate rating', 'Rating color', 'Rating text']].value_counts().reset_index()
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:"Rating Count"})
```

```python
ratings
```

## Observation
1.When Rating is between 4.5 to 4.9---> Excellent

2.When Rating are between 4.0 to 4.4--->very good

3.when Rating is between 3.5 to 3.9----> good

4.when Rating is between 3.0 to 3.4----> average

5.when Rating is between 2.5 to 2.9----> average

6.when Rating is between 2.0 to 2.4----> Poor


```python
## Bar Plot
sns.barplot(data=ratings,x='Aggregate rating',y='Rating Count')
```

```python
sns.barplot(data=ratings,x='Aggregate rating',y='Rating Count',hue='Rating color',palette=['blue','red','orange','yellow','green','green'])
```

Observation:

Not Rated count is very high
Maximum number of rating are between 2.5 to 3.4


```python
sns.countplot(data=ratings,x='Rating color')
```
